```python
import pandas as pd

# Define the file path
file_path = '/content/drive/MyDrive/Tugas/TEKS/CSV/youtube_comments.csv'

# Read the CSV file into a pandas DataFrame
try:
    saved_comments_df = pd.read_csv(file_path)
    print(f"Successfully read data from {file_path}")
    display(saved_comments_df)

except FileNotFoundError:
    print(f"Error: The file {file_path} was not found.")
except Exception as e:
    print(f"An error occurred while reading the CSV file: {e}")
```

```python
import re
import nltk
from nltk.corpus import stopwords

# Download necessary NLTK data
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')
try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')
```

```python
# 1. Case Folding (already done in cell 6_wq8Wqc1q4A, but including for completeness)
saved_comments_df['text'] = saved_comments_df['text'].str.lower()
print("Preprocessing steps applied: Case Folding.")
display(saved_comments_df.head(100))
```

```python
# 2. Removing Punctuation
def remove_punctuation(text):
    # Ensure text is a string before applying lower()
    if not isinstance(text, str):
        return text
    return re.sub(r'[^\w\s]', '', text.lower()) # Apply lower() here for consistency

saved_comments_df['text'] = saved_comments_df['text'].apply(remove_punctuation)
print("Preprocessing steps applied: Case Folding, Removing Punctuation.")
display(saved_comments_df.head(100))
```

```python
# 3. Tokenization and Removing Stop Words
stop_words = set(stopwords.words('indonesian'))

def tokenize_and_filter_stopwords(text):
    # Ensure text is a string
    if not isinstance(text, str):
        return []
    # Tokenize
    tokens = text.split() # Simple split after removing punctuation
    # Remove stop words
    filtered_tokens = [word for word in tokens if word not in stop_words]
    return filtered_tokens

saved_comments_df['text_processed'] = saved_comments_df['text'].apply(tokenize_and_filter_stopwords)
print("Preprocessing steps applied: Case Folding, Removing Punctuation, Tokenization, and Removing Stop Words.")
display(saved_comments_df.head(100))
```

```python
# Import the StemmerFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Function to perform stemming on a list of tokens
def stem_tokens(tokens):
    # Ensure tokens is a list
    if not isinstance(tokens, list):
        return []
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    return stemmed_tokens

# Apply stemming to the 'text_processed' column
saved_comments_df['text_stemmed'] = saved_comments_df['text_processed'].apply(stem_tokens)

print("Preprocessing steps applied: Case Folding, Removing Punctuation, Tokenization, Removing Stop Words, and Stemming.")
display(saved_comments_df.head(100))
```

```python
import nltk
from nltk.corpus import stopwords
from collections import Counter
import re

# Download necessary NLTK data (if you haven't already)
nltk.download('punkt')
nltk.download('stopwords')

# Get the list of Indonesian stop words
stop_words = set(stopwords.words('indonesian'))

# Function to tokenize and remove stop words
def tokenize_and_filter(text):
    # Remove non-alphanumeric characters and tokenize
    tokens = re.findall(r'\b\w+\b', text.lower())
    # Remove stop words
    filtered_tokens = [word for word in tokens if word not in stop_words]
    return filtered_tokens

# Apply the function to the 'text' column and get a list of all tokens
all_tokens = [token for text in saved_comments_df['text'] for token in tokenize_and_filter(text)]

# Count the frequency of each token
word_counts = Counter(all_tokens)

# Get the most common words
most_common_words = word_counts.most_common(20) # Get the top 20

print("Most common words (after removing stop words):")
for word, count in most_common_words:
    print(f"{word}: {count}")
```
