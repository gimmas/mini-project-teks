```python
# Import necessary libraries
from googleapiclient.discovery import build
import pandas as pd
```

```python
# Replace with your API key and video ID
api_key = "AIzaSyBZ37eeoC9-bZ5iuXsp_aqaJhQ7tgSUeYE"
video_id = "P1NbOV7avUs"
max_results = 10000 # Desired number of comments

# Build the YouTube Data API client
youtube = build('youtube', 'v3', developerKey=api_key)

comments = []
next_page_token = None

while len(comments) < max_results:
    try:
        # Get comments for the video
        results = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=min(100, max_results - len(comments)),
            pageToken=next_page_token
        ).execute()

        for item in results["items"]:
            comment = item["snippet"]["topLevelComment"]["snippet"]
            comments.append({
                'author': comment["authorDisplayName"],
                'published_at': comment["publishedAt"],
                'text': comment["textDisplay"]
            })

        next_page_token = results.get('nextPageToken')

        if not next_page_token:
            break # No more pages

    except Exception as e:
        print(f"An error occurred: {e}")
        break # Exit loop if an error occurs

print(f"Successfully crawled {len(comments)} comments.")

# You can now convert the comments list to a pandas DataFrame or save it to a file.
comments_df = pd.DataFrame(comments)
display(comments_df.head())
```
