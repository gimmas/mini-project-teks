```python
from transformers import pipeline

# Load a pre-trained sentiment analysis model for Indonesian
sentiment_analyzer = pipeline("sentiment-analysis", model="w11wo/indonesian-roberta-base-sentiment-classifier")

print("Sentiment analysis model loaded.")
```

```python
import pandas as pd

# Function to get sentiment label and confidence from the model output
def get_sentiment_and_confidence(text):
    try:
        # Ensure text is a string; handle non-string types or NaN
        if not isinstance(text, str) or pd.isna(text):
            return 'netral', 0.0

        result = sentiment_analyzer(text)[0] # Get the first prediction
        label = result['label']
        score = result['score']

        # Map labels to 'positif', 'negatif', 'netral'
        if label == 'positive':
            return 'positif', score
        elif label == 'negative':
            return 'negatif', score
        else: # Handle neutral or other labels if the model produces them
            return 'netral', score
    except Exception as e:
        # print(f"Error processing text: {text} - {e}") # Uncomment for debugging
        return 'netral', 0.0 # Assign 'netral' and 0.0 for any errors

# Apply the sentiment analysis to the 'comment_text' column
print("Applying sentiment analysis to comments and generating pseudo-labels...")

# Apply the function and expand the results into two new columns
set_2_df[['pseudo_label', 'confidence_score']] = set_2_df['comment_text'].apply(lambda x: pd.Series(get_sentiment_and_confidence(x)))

print("Pseudo-labeling complete!")

print("\nFirst 5 rows of set_2_df with pseudo-labels and confidence scores:")
display(set_2_df.head())

print("\nPseudo-label distribution:")
display(set_2_df['pseudo_label'].value_counts())
```

```python
confidence_threshold = 0.9

# Filter for high-confidence pseudo-labels
high_confidence_pseudo_labels = set_2_df[set_2_df['confidence_score'] >= confidence_threshold]

print(f"Original number of pseudo-labeled comments: {len(set_2_df)}")
print(f"Number of high-confidence pseudo-labeled comments (confidence >= {confidence_threshold}): {len(high_confidence_pseudo_labels)}")

print("\nFirst 5 rows of high-confidence pseudo-labeled data:")
display(high_confidence_pseudo_labels.head())

print("\nDistribution of pseudo-labels in high-confidence data:")
display(high_confidence_pseudo_labels['pseudo_label'].value_counts())
```

```python
print("First 5 rows of the high_confidence_pseudo_labels DataFrame:")
display(high_confidence_pseudo_labels.head())

print("\nShape of high_confidence_pseudo_labels (rows, columns):")
print(high_confidence_pseudo_labels.shape)

print("\nDistribution of pseudo_labels in high-confidence data:")
display(high_confidence_pseudo_labels['pseudo_label'].value_counts())
```

```python
import pandas as pd

# 1. Create a new DataFrame named final_labeled_dataset by copying high_confidence_pseudo_labels
#    and drop the 'label' column as 'pseudo_label' is the intended label.
final_labeled_dataset = high_confidence_pseudo_labels.drop(columns=['label']).copy()

print("Final Labeled Dataset created and 'label' column dropped.")

# 2. Display the first 5 rows of final_labeled_dataset
print("\nFirst 5 rows of final_labeled_dataset:")
display(final_labeled_dataset.head())

# 3. Print the shape (number of rows and columns) of final_labeled_dataset
print(f"\nShape of final_labeled_dataset: {final_labeled_dataset.shape}")

# 4. Display the value counts of the 'pseudo_label' column in final_labeled_dataset
print("\nDistribution of 'pseudo_label' in final_labeled_dataset:")
display(final_labeled_dataset['pseudo_label'].value_counts())
```
